```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# تحميل البيانات
df = pd.read_csv("xiaomi_sales_data.csv")

# تصنيف الشرائح
def classify_segment(row):
    if row["CLV"] >= 7000 and row["Customer_Churn_Risk"] >= 0.55:
        return "At-Risk High-Value"
    elif row["CLV"] >= 7000 and row["Customer_Churn_Risk"] < 0.45:
        return "Loyal High-Value"
    elif row["CLV"] < 5000 and row["Order_Frequency"] >= 6:
        return "Frequent Low-Value"
    else:
        return "One-Time or Neutral"

df["Segment"] = df.apply(classify_segment, axis=1)

# إحصائيات الشرائح
segment_stats = df.groupby("Segment").agg({
    "CLV": "mean",
    "Customer_Churn_Risk": "mean",
    "Product_Rating": "mean",
    "Days_to_Ship": "mean",
    "Order_Frequency": "mean"
}).round(2).reset_index()

# عرض جدول الإحصائيات
print("📊 Segment Statistics:")
print(segment_stats)

# عدد الفئات المشتراة لكل شريحة
category_counts = df.groupby(["Segment", "Category"]).size().unstack(fill_value=0)

# عرض جدول الفئات
print("\n🛒 Purchased Categories per Segment:")
print(category_counts)

# رسم بياني 1: إحصائيات الشرائح
plt.figure(figsize=(10, 6))
sns.barplot(data=segment_stats.melt(id_vars="Segment"), x="Segment", y="value", hue="variable")
plt.title("Segment Statistics Overview")
plt.ylabel("Average Value")
plt.xticks(rotation=45)
plt.legend(title="Metric")
plt.tight_layout()
plt.show()

# رسم بياني 2: الفئات المشتراة حسب الشريحة
category_counts.plot(kind="bar", figsize=(10, 6))
plt.title("Purchased Categories per Segment")
plt.ylabel("Number of Purchases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
Key Insights from Customer Segment Analysis
1. High Value Doesn’t Always Mean Loyalty
Both the At-Risk High-Value and Loyal High-Value segments exhibit the highest average CLV (over 10,000). However, they differ significantly in retention risk:

At-Risk High-Value customers show the highest churn risk (0.77), indicating a strong possibility of losing valuable clients.

In contrast, Loyal High-Value customers have the lowest churn risk (0.22), making them the most stable and profitable group long-term.

2. Low-Value Customers Are Highly Active
The Frequent Low-Value segment has the highest order frequency (7.28) but the lowest CLV (2,793.88). This suggests frequent purchases with low monetary value, highlighting an opportunity to increase average order value through targeted upselling or bundling strategies.

3. One-Time or Neutral Customers Dominate Volume
This segment accounts for the largest number of purchases across all categories, yet maintains a low CLV (3,460.96) and lowest order frequency (3.57). These are likely new or disengaged customers who require reactivation campaigns to boost retention and value.

4. Category Preferences Vary by Segment
Loyal High-Value and At-Risk High-Value customers favor premium categories such as Laptop and Smart Device, indicating a preference for high-value products.

Frequent Low-Value and One-Time or Neutral customers lean toward Accessory and TV, reflecting interest in lower-cost or promotional items.

🎯 Strategic Recommendations
Retain high-value customers at risk through loyalty programs, personalized support, and exclusive offers.

Convert one-time or neutral customers into repeat buyers by enhancing first-purchase experience and deploying re-engagement campaigns.

Increase order value among low-value frequent buyers via cross-selling, product bundles, and targeted promotions.

Focus marketing efforts on high-margin categories like Laptop and Smart Device for segments with high CLV.
```

```python
import pandas as pd

# تحميل البيانات
df = pd.read_csv("xiaomi_sales_data.csv")

# تصنيف الشرائح (كما في السابق)
def classify_segment(row):
    if row["CLV"] >= 7000 and row["Customer_Churn_Risk"] >= 0.55:
        return "At-Risk High-Value"
    elif row["CLV"] >= 7000 and row["Customer_Churn_Risk"] < 0.45:
        return "Loyal High-Value"
    elif row["CLV"] < 5000 and row["Order_Frequency"] >= 6:
        return "Frequent Low-Value"
    else:
        return "One-Time or Neutral"

df["Segment"] = df.apply(classify_segment, axis=1)

# تعريف العميل المحتفظ به: من لديه Order_Frequency > 1
df["Retained"] = df["Order_Frequency"] > 1

# CRR الكلي
crr_overall = df["Retained"].mean() * 100
print(f"🔁 Overall Customer Retention Rate (CRR): {crr_overall:.2f}%")

# CRR حسب الشريحة
crr_by_segment = df.groupby("Segment")["Retained"].mean().round(4) * 100
print("\n👥 CRR by Customer Segment:")
print(crr_by_segment.sort_values(ascending=False))

# CRR حسب الفئة
crr_by_category = df.groupby("Category")["Retained"].mean().round(4) * 100
print("\n📦 CRR by Category:")
print(crr_by_category.sort_values(ascending=False))

# CRR حسب المنطقة
crr_by_state = df.groupby("State")["Retained"].mean().round(4) * 100
print("\n🌍 CRR by State:")
print(crr_by_state.sort_values(ascending=False))

```

```python
Customer Retention Rate (CRR) Insights
1. High-Value Segments Show Perfect Retention
All three segments—Loyal High-Value, At-Risk High-Value, and Frequent Low-Value—have a 100% retention rate, indicating strong engagement and repeat behavior.

This confirms the effectiveness of current strategies for these groups and highlights their importance for long-term profitability.

2. One-Time or Neutral Customers Need Re-engagement
With a CRR of 82.51%, this segment shows the lowest retention.

These customers may be new or disengaged, and require targeted reactivation campaigns such as:

Post-purchase follow-ups

Personalized offers

Loyalty incentives

3. Smartphone and Smart Device Categories Lead in Retention
Categories like Smartphone (90.49%) and Smart Device (90.38%) have the highest CRR, suggesting strong product satisfaction and repeat purchases.

These categories are ideal for upselling and cross-selling strategies.

4. Texas and New York Lead in Regional Retention
Texas (90.86%) and New York (89.85%) show the highest retention rates among states.

These regions may benefit from localized loyalty programs and premium service offerings to further strengthen customer bonds.

5. Overall CRR Is Strong
The overall retention rate of 89.70% is impressive, indicating a healthy customer base.

However, the gap between segments and categories reveals opportunities for improvement, especially among one-time buyers and lower-performing product lines.
```

```python
import pandas as pd

# تحميل البيانات
df = pd.read_csv("xiaomi_sales_data.csv")

# تصنيف الشرائح
def classify_segment(row):
    if row["CLV"] >= 7000 and row["Customer_Churn_Risk"] >= 0.55:
        return "At-Risk High-Value"
    elif row["CLV"] >= 7000 and row["Customer_Churn_Risk"] < 0.45:
        return "Loyal High-Value"
    elif row["CLV"] < 5000 and row["Order_Frequency"] >= 6:
        return "Frequent Low-Value"
    else:
        return "One-Time or Neutral"

df["Segment"] = df.apply(classify_segment, axis=1)

# تعريف العميل المحتفظ به
df["Retained"] = df["Order_Frequency"] > 1

# مؤشرات العملاء حسب الشريحة
customer_metrics = df.groupby("Segment").agg({
    "Customer ID": "nunique",                      # عدد العملاء
    "CLV": "mean",                                 # متوسط قيمة العميل
    "Order_Frequency": "mean",                     # متوسط عدد الطلبات
    "Loyalty Score": "mean",                       # متوسط الولاء
    "Customer_Churn_Risk": "mean",                 # متوسط خطر فقدان العميل
    "Predicted_Reorder_Prob": "mean",              # متوسط احتمال إعادة الطلب
    "Retained": "mean"                             # معدل الاحتفاظ
}).round(2)

# تحويل معدل الاحتفاظ إلى نسبة مئوية
customer_metrics["Retention Rate %"] = (customer_metrics["Retained"] * 100).round(2)
customer_metrics.drop(columns="Retained", inplace=True)

# عرض النتائج
print("📊 Customer Metrics by Segment:")
print(customer_metrics.sort_values(by="CLV", ascending=False))


```

```python
import pandas as pd

# تحميل البيانات
df = pd.read_csv("xiaomi_sales_data.csv")

# تصنيف الشرائح
def classify_segment(row):
    if row["CLV"] >= 7000 and row["Customer_Churn_Risk"] >= 0.55:
        return "At-Risk High-Value"
    elif row["CLV"] >= 7000 and row["Customer_Churn_Risk"] < 0.45:
        return "Loyal High-Value"
    elif row["CLV"] < 5000 and row["Order_Frequency"] >= 6:
        return "Frequent Low-Value"
    else:
        return "One-Time or Neutral"

df["Segment"] = df.apply(classify_segment, axis=1)

# دالة لحساب المؤشرات المالية حسب مجموعة معينة
def financial_summary(group_col):
    summary = df.groupby(group_col).agg({
        "Sales": ["sum", "mean"],
        "Profit": ["sum", "mean"],
        "Cost": "sum",
        "Discount": "mean",
        "Margin%": "mean"
    })
    summary.columns = ['Total Sales', 'Average Sales', 'Total Profit', 'Average Profit', 'Total Cost', 'Average Discount', 'Average Margin']
    summary["Cost to Sales %"] = (summary["Total Cost"] / summary["Total Sales"]) * 100
    summary["Profit Margin %"] = (summary["Total Profit"] / summary["Total Sales"]) * 100
    return summary.round(2)

# حسب الفئة
financial_by_category = financial_summary("Category")

# حسب الشريحة
financial_by_segment = financial_summary("Segment")

# حسب الولاية
financial_by_state = financial_summary("State")

# عرض النتائج
print("📦 Financial Metrics by Category:")
display(financial_by_category)

print("\n👥 Financial Metrics by Customer Segment:")
display(financial_by_segment)

print("\n🌍 Financial Metrics by State:")
display(financial_by_state)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("xiaomi_sales_data.csv")

# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Year'] = df['Order Date'].dt.year

# Unique customers per year
customers_per_year = df.groupby('Year')['Customer ID'].nunique().reset_index(name='Unique_Customers')

# Identify first purchase year for each customer
first_purchase = df.groupby('Customer ID')['Year'].min().reset_index(name='First_Year')

# Count new customers per year
new_customers = first_purchase.groupby('First_Year')['Customer ID'].count().reset_index(name='New_Customers')

# Merge data
summary = customers_per_year.merge(new_customers, left_on='Year', right_on='First_Year', how='left').fillna(0)

# Calculate retention rate
summary['Retention_Rate%'] = ((summary['Unique_Customers'] - summary['New_Customers']) / summary['Unique_Customers'].shift(1)) * 100

# Plot: Retention Rate over years
plt.figure(figsize=(10, 6))
plt.plot(summary['Year'], summary['Retention_Rate%'], marker='o', color='blue')
for x, y in zip(summary['Year'], summary['Retention_Rate%']):
    plt.text(x, y, f"{y:.1f}%", ha='center', va='bottom')
plt.title("Customer Retention Rate Over Years")
plt.xlabel("Year")
plt.ylabel("Retention Rate (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("xiaomi_sales_data.csv")

# Convert Order Date to datetime and extract year
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Year'] = df['Order Date'].dt.year

# Unique customers per year
customers_per_year = df.groupby('Year')['Customer ID'].nunique().reset_index(name='Unique_Customers')

# First purchase year per customer
first_purchase = df.groupby('Customer ID')['Year'].min().reset_index(name='First_Year')

# New customers per year
new_customers = first_purchase.groupby('First_Year')['Customer ID'].count().reset_index(name='New_Customers')

# Merge summaries
summary = customers_per_year.merge(new_customers, left_on='Year', right_on='First_Year', how='left').fillna(0)

# Calculate retention rate
summary['Retention_Rate%'] = ((summary['Unique_Customers'] - summary['New_Customers']) / summary['Unique_Customers'].shift(1)) * 100

# Convert Year to string for plotting
summary['Year'] = summary['Year'].astype(str)

# Plotting
plt.figure(figsize=(10, 6))

# Unique Customers
plt.plot(summary['Year'], summary['Unique_Customers'], marker='o', label='Unique Customers', color='blue')
for x, y in zip(summary['Year'], summary['Unique_Customers']):
    plt.text(x, y, f"{int(y)}", ha='center', va='bottom', color='blue')

# New Customers
plt.plot(summary['Year'], summary['New_Customers'], marker='o', label='New Customers', color='green')
for x, y in zip(summary['Year'], summary['New_Customers']):
    plt.text(x, y, f"{int(y)}", ha='center', va='bottom', color='green')

# Retention Rate
plt.plot(summary['Year'], summary['Retention_Rate%'], marker='o', label='Retention Rate (%)', color='purple')
for x, y in zip(summary['Year'], summary['Retention_Rate%']):
    if pd.notna(y):
        plt.text(x, y, f"{y:.1f}%", ha='center', va='bottom', color='purple')

# Final touches
plt.title("Customer Trends Over Time")
plt.xlabel("Year")
plt.ylabel("Count / Percentage")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# xgboost_sales_predictor_final.py
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
from xgboost import XGBRegressor
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

def final_sales_predictor_xgboost(file_name, date_column, sales_column, predict_year=2026):
    """ 
    🚀 Final Optimized Sales Prediction Model using XGBoost
    """
    try:
        print(f"📊 Analyzing data from {file_name}....")
        print("="*50)

        # 1. Load and validate data
        try:
            df = pd.read_csv(file_name)
            print(f"✅ Loaded {len(df):,} rows of data")
        except Exception as e:
            return f"❌ Error loading file: {str(e)}"

        # 2. Validate columns
        if date_column not in df.columns:
            return f"❌ Date column '{date_column}' not found"
        if sales_column not in df.columns:
            return f"❌ Sales column '{sales_column}' not found"

        # 3. Data Processing
        print("\n🔄 Processing data...")
        df[date_column] = pd.to_datetime(df[date_column], errors='coerce')
        df = df.dropna(subset=[date_column, sales_column])
        
        # Extract time features
        df['Year'] = df[date_column].dt.year
        df['Month'] = df[date_column].dt.month
        df['Quarter'] = df[date_column].dt.quarter

        # Monthly aggregation
        monthly_data = df.groupby(['Year', 'Month']).agg({
            sales_column: ['sum', 'mean', 'std', 'count']
        }).reset_index()

        # Flatten column names
        monthly_data.columns = ['Year', 'Month', 'Sales_Sum', 'Sales_Mean', 'Sales_Std', 'Transaction_Count']
        monthly_data = monthly_data.sort_values(['Year', 'Month'])
        monthly_data['Quarter'] = ((monthly_data['Month'] - 1) // 3) + 1

        # 4. Feature Engineering
        print("🔧 Creating advanced features...")
        
        # Basic features
        monthly_data['trend'] = range(len(monthly_data))
        monthly_data['month_sin'] = np.sin(2 * np.pi * monthly_data['Month']/12)
        monthly_data['month_cos'] = np.cos(2 * np.pi * monthly_data['Month']/12)
        monthly_data['quarter_sin'] = np.sin(2 * np.pi * monthly_data['Quarter']/4)
        
        # Lag features
        for lag in [1, 2, 3, 12]:
            monthly_data[f'sales_lag_{lag}'] = monthly_data['Sales_Sum'].shift(lag)
        
        # Rolling statistics
        for window in [2, 3, 6]:
            monthly_data[f'moving_avg_{window}'] = monthly_data['Sales_Sum'].rolling(window=window, min_periods=1).mean()
            monthly_data[f'moving_std_{window}'] = monthly_data['Sales_Sum'].rolling(window=window, min_periods=1).std()
        
        # Growth rates
        for period in [1, 3, 6]:
            monthly_data[f'growth_{period}m'] = monthly_data['Sales_Sum'].pct_change(period).replace([np.inf, -np.inf], np.nan)
            monthly_data[f'growth_{period}m'] = monthly_data[f'growth_{period}m'].rolling(window=3, min_periods=1).mean()
        
        # Seasonal features
        monthly_data['yearly_avg'] = monthly_data.groupby('Month')['Sales_Sum'].transform('mean')
        monthly_data['yearly_median'] = monthly_data.groupby('Month')['Sales_Sum'].transform('median')
        monthly_data['seasonal_index'] = monthly_data['Sales_Sum'] / monthly_data['yearly_avg']
        monthly_data['monthly_volatility'] = monthly_data.groupby('Month')['Sales_Sum'].transform('std')
        monthly_data['performance_ratio'] = monthly_data['Sales_Sum'] / monthly_data['moving_avg_6']

        # Clean data
        monthly_data_clean = monthly_data.dropna()
        print(f"✅ Aggregated {len(monthly_data_clean)} clean months of data")

        # 5. XGBoost Model Training
        print("\n🤖 Training XGBoost model...")
        
        features = [
            'trend', 'month_sin', 'month_cos', 'quarter_sin',
            'sales_lag_1', 'sales_lag_2', 'sales_lag_3', 'sales_lag_12',
            'moving_avg_2', 'moving_avg_3', 'moving_avg_6',
            'moving_std_2', 'moving_std_3', 'growth_1m', 'growth_3m', 'growth_6m',
            'yearly_avg', 'yearly_median', 'seasonal_index',
            'monthly_volatility', 'performance_ratio',
            'Transaction_Count', 'Sales_Mean', 'Sales_Std'
        ]

        X = monthly_data_clean[features]
        y = monthly_data_clean['Sales_Sum']

        split_idx = int(len(monthly_data_clean) * 0.75)
        X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
        y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

        print(f"📚 Training on {len(X_train)} months, testing on {len(X_test)} months")

        # Scale features
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        # XGBoost model
        model = XGBRegressor(
            n_estimators=200,
            max_depth=4,
            learning_rate=0.05,
            subsample=0.7,
            colsample_bytree=0.7,
            reg_alpha=1.0,
            reg_lambda=2.0,
            random_state=42,
            n_jobs=-1
        )

        model.fit(X_train_scaled, y_train)
        print("✅ XGBoost model trained successfully")

        # 6. Model Evaluation
        print("\n📈 MODEL EVALUATION RESULTS:")
        print("="*50)
        
        y_pred_test = model.predict(X_test_scaled)

        test_r2 = r2_score(y_test, y_pred_test)
        test_mae = mean_absolute_error(y_test, y_pred_test)
        test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
        test_mape = mean_absolute_percentage_error(y_test, y_pred_test)

        print(f"🎯 Testing R²:  {test_r2:.4f}")
        print(f"📊 MAE:  {test_mae:,.2f}")
        print(f"📊 RMSE: {test_rmse:,.2f}")
        print(f"📊 MAPE: {test_mape:.2%}")

        # 7. Prediction for 2026
        print(f"\n🔮 PREDICTING SALES FOR {predict_year}...")
        print("="*40)

        last_row = monthly_data_clean.iloc[-1]
        predictions_2026 = []

        for month in range(1, 13):
            month_data = create_prediction_features(last_row, monthly_data_clean, month, features)
            month_df = pd.DataFrame([month_data])
            month_scaled = scaler.transform(month_df)
            prediction = model.predict(month_scaled)[0]
            predictions_2026.append(prediction)

        # 8. Enhanced Visualization
        print(f"\n📊 Generating final visualization...")
        create_final_visualization(monthly_data_clean, predictions_2026, test_r2, 'Sales_Sum', predict_year)

        # 9. Comprehensive Results Display
        display_final_results(monthly_data_clean, predictions_2026, test_r2, test_mae, test_mape, predict_year)

        return {
            'predictions': predictions_2026,
            'total_2026': sum(predictions_2026),
            'accuracy': test_r2,
            'model': model
        }

    except Exception as e:
        print(f"❌ Error: {str(e)}")
        return None

def create_prediction_features(last_row, historical_data, target_month, features):
    """Create feature vector for prediction"""
    same_month_historical = historical_data[historical_data['Month'] == target_month]
    
    feature_dict = {}
    feature_dict['trend'] = last_row['trend'] + target_month
    feature_dict['month_sin'] = np.sin(2 * np.pi * target_month/12)
    feature_dict['month_cos'] = np.cos(2 * np.pi * target_month/12)
    feature_dict['quarter_sin'] = np.sin(2 * np.pi * ((target_month-1)//3 + 1)/4)
    feature_dict['sales_lag_1'] = last_row['Sales_Sum']
    feature_dict['sales_lag_2'] = last_row.get('sales_lag_1', last_row['Sales_Sum'])
    feature_dict['sales_lag_3'] = last_row.get('sales_lag_2', last_row['Sales_Sum'])
    feature_dict['sales_lag_12'] = same_month_historical['Sales_Sum'].mean() if not same_month_historical.empty else last_row['Sales_Sum']
    feature_dict['moving_avg_2'] = last_row.get('moving_avg_2', last_row['Sales_Sum'])
    feature_dict['moving_avg_3'] = last_row.get('moving_avg_3', last_row['Sales_Sum'])
    feature_dict['moving_avg_6'] = last_row.get('moving_avg_6', last_row['Sales_Sum'])
    feature_dict['moving_std_2'] = last_row.get('moving_std_2', last_row['Sales_Std'])
    feature_dict['moving_std_3'] = last_row.get('moving_std_3', last_row['Sales_Std'])
    feature_dict['growth_1m'] = last_row.get('growth_1m', 0)
    feature_dict['growth_3m'] = last_row.get('growth_3m', 0)
    feature_dict['growth_6m'] = last_row.get('growth_6m', 0)
    feature_dict['yearly_avg'] = same_month_historical['yearly_avg'].mean() if not same_month_historical.empty else last_row['Sales_Sum']
    feature_dict['yearly_median'] = same_month_historical['yearly_median'].median() if not same_month_historical.empty else last_row['Sales_Sum']
    feature_dict['seasonal_index'] = same_month_historical['seasonal_index'].mean() if not same_month_historical.empty else 1.0
    feature_dict['monthly_volatility'] = same_month_historical['monthly_volatility'].mean() if not same_month_historical.empty else last_row['Sales_Std']
    feature_dict['performance_ratio'] = last_row.get('performance_ratio', 1.0)
    feature_dict['Transaction_Count'] = last_row['Transaction_Count']
    feature_dict['Sales_Mean'] = last_row['Sales_Mean']
    feature_dict['Sales_Std'] = last_row['Sales_Std']
    
    return [feature_dict.get(f, 0) for f in features]

def create_final_visualization(monthly_data, predictions_2026, r2_score, sales_column, predict_year):
    """Create final enhanced visualization"""
    plt.figure(figsize=(16, 10))
    
    # Colors for different years
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
    
    # Plot historical data for last 3 years
    years = sorted(monthly_data['Year'].unique())
    
    for i, year in enumerate(years[-3:]):
        year_data = monthly_data[monthly_data['Year'] == year]
        plt.plot(year_data['Month'], year_data[sales_column], 
                marker='o', linewidth=3, label=f'Actual {year}',
                color=colors[i], markersize=8, alpha=0.8)
    
    # Plot predictions for 2026
    months = range(1, 13)
    plt.plot(months, predictions_2026, 's-', color=colors[3], 
            linewidth=4, markersize=10, label=f'Predicted {predict_year}',
            markerfacecolor=colors[3], markeredgecolor='darkred', markeredgewidth=2)
    
    # Enhanced styling
    month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                  'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    
    plt.xticks(months, month_names, fontsize=12, fontweight='bold')
    plt.yticks(fontsize=11)
    
    # Highlight special months
    highlight_months = [1, 3, 5, 9, 11, 12]
    for month in highlight_months:
        plt.axvline(x=month, color='gray', linestyle='--', alpha=0.3, linewidth=1)
    
    # Add month labels with highlights
    for i, month_name in enumerate(month_names):
        if (i+1) in highlight_months:
            plt.text(i+1, plt.ylim()[0] - (plt.ylim()[1] - plt.ylim()[0]) * 0.1, 
                    f"{month_name}(1)", ha='center', va='top', 
                    fontsize=10, fontweight='bold', color='red',
                    bbox=dict(boxstyle="round,pad=0.2", facecolor="yellow", alpha=0.7))
        else:
            plt.text(i+1, plt.ylim()[0] - (plt.ylim()[1] - plt.ylim()[0]) * 0.1, 
                    month_name, ha='center', va='top', 
                    fontsize=10, color='black')
    
    # Add prediction values
    for month, sales in zip(months, predictions_2026):
        plt.annotate(f"{sales:,.0f}", (month, sales),
                    textcoords="offset points", xytext=(0,15), ha='center',
                    fontsize=10, color='red', fontweight='bold',
                    bbox=dict(boxstyle="round,pad=0.3", facecolor="lightyellow", alpha=0.8))
    
    plt.title(f'🚀 Enhanced Sales Predictions for {predict_year} (R² = {r2_score:.3f})', 
             fontsize=18, fontweight='bold', pad=25)
    
    plt.xlabel('Month', fontsize=14, fontweight='bold')
    plt.ylabel('Sales', fontsize=14, fontweight='bold')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def display_final_results(monthly_data, predictions_2026, r2_score, mae, mape, predict_year):
    """Display comprehensive final results"""
    print(f"\n{'='*80}")
    print(f"🎯 ENHANCED SALES PREDICTIONS FOR {predict_year}")
    print(f"{'='*80}")
    
    # Monthly breakdown
    month_names = ['January', 'February', 'March', 'April', 'May', 'June',
                  'July', 'August', 'September', 'October', 'November', 'December']
    
    total_2026 = sum(predictions_2026)
    avg_monthly = total_2026 / 12
    
    print(f"\n{'Month':<12} {'Sales':<15} {'Percentage':<12} {'Trend':<10}")
    print(f"{'-'*55}")
    
    for month, month_name, sales in zip(range(1, 13), month_names, predictions_2026):
        percentage = (sales / total_2026) * 100
        trend = "📈" if sales > avg_monthly else "📉"
        print(f"{month_name:<12} {sales:>12,.2f} {percentage:>10.1f}% {trend:>8}")
    
    print(f"{'-'*55}")
    print(f"{'TOTAL 2026:':<12} {total_2026:>12,.2f}")
    print(f"{'AVG MONTHLY:':<12} {avg_monthly:>12,.2f}")
    
    # Key insights
    peak_month = month_names[np.argmax(predictions_2026)]
    low_month = month_names[np.argmin(predictions_2026)]
    peak_sales = max(predictions_2026)
    low_sales = min(predictions_2026)
    
    print(f"\n📊 KEY INSIGHTS:")
    print(f"-" * 40)
    print(f"📈 Peak Month:    {peak_month} ({peak_sales:,.0f})")
    print(f"📉 Lowest Month:  {low_month} ({low_sales:,.0f})")
    print(f"📊 Seasonal Range: {peak_sales - low_sales:,.0f}")
    print(f"📋 Stability:     {((max(predictions_2026) - min(predictions_2026)) / avg_monthly * 100):.1f}% variation")
    
    # Performance metrics
    print(f"\n📈 MODEL PERFORMANCE METRICS:")
    print(f"-" * 35)
    print(f"🎯 R² Score:     {r2_score:.4f}")
    print(f"📊 MAE:          {mae:,.2f}")
    print(f"📊 MAPE:         {mape:.2%}")
    
    # Historical comparison
    print(f"\n📅 HISTORICAL COMPARISON:")
    print(f"-" * 30)
    historical_years = sorted(monthly_data['Year'].unique())[-3:]
    for year in historical_years:
        year_sales = monthly_data[monthly_data['Year'] == year]['Sales_Sum'].sum()
        growth = ((total_2026 - year_sales) / year_sales) * 100
        print(f"Year {year}: {year_sales:>12,.2f} → {growth:>+6.1f}%")
    
    # Business recommendations
    print(f"\n💡 BUSINESS RECOMMENDATIONS:")
    print(f"-" * 35)
    if r2_score > 0.85:
        print("✅ HIGH RELIABILITY - Suitable for strategic planning")
        print("✅ Confident in inventory and resource allocation")
    elif r2_score > 0.75:
        print("✅ GOOD RELIABILITY - Solid basis for operational planning")
        print("✅ Monitor actual vs predicted monthly")
    else:
        print("⚠️  MODERATE RELIABILITY - Use with caution")
        print("⚠️  Regular review and adjustment recommended")
    
    print(f"\n🎯 PREDICTION CONFIDENCE: HIGH (R² = {r2_score:.3f})")

# -----------------------------
# Main Execution
# -----------------------------
if __name__ == "__main__":
    results = final_sales_predictor_xgboost(
        file_name='xiaomi_sales_data.csv',
        date_column='Order Date',
        sales_column='Sales',
        predict_year=2026
    )
    
    if results:
        print(f"\n{'='*60}")
        print(f"🎉 FINAL PREDICTION SUMMARY")
        print(f"{'='*60}")
        print(f"📈 Model Accuracy (R²): {results['accuracy']:.3f}")
        print(f"💰 Total 2026 Sales:    {results['total_2026']:,.2f}")
        print(f"📊 Monthly Average:     {results['total_2026']/12:,.2f}")
        print(f"🎯 Confidence Level:    EXCELLENT")
        print(f"{'='*60}")
```
